```python
import pandas as pd
from textgenrnn import textgenrnn

textgen = textgenrnn()
```

```python
# load text
raw_text = pd.read_csv("dickens.csv", engine="python").dropna()
raw_text = raw_text.drop(["title"], axis=1)
raw_text.to_csv("raw.csv",index=False)

# Check text
raw_text
```

```python
# Train model
textgen.train_from_file('raw.csv',
                        num_epochs=1,
                        is_csv=True)
# Save model
textgen.save(weights_path='low expectations 1 full epoch.hdf5')
```

```python
# Load model
low_expectations = textgenrnn('low expectations 1 full epoch.hdf5')
```

```python
# Generate sentences
texts = low_expectations.generate(1000,
                                  temperature = 0.5,
                                  return_as_list = True,
                                  progress = True)
# Save to CSV
df_texts = pd.DataFrame(texts, columns=["Line"])
df_texts.to_csv("generated lines.csv", index = False)
```
